```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr
wr.filterwarnings('ignore')
```

Data link: https://media.geeksforgeeks.org/wp-content/uploads/20250731152112425017/WineQT.csv

```python
# create panda's dataframe with csv data
df = pd.read_csv('WinteQT.csv')
print(df.head)
```

```python
# analyze data
print(f'shape of data = {df.shape}')
print('-------------------------------------------') # separate data
df.info()
print('-------------------------------------------')
print(f'data summary: {df.describe().T}') # transpose to show data as rows
```

```python
# convert column names into a list for ease of manipulation
df.columns.tolist()
```

```python
# check missing values
df.isnull().sum()
```

```python
# check for duplicate values
df.nunique()
```

Univariate analysis is used in plotting the data

```python
# count of wine with quality rating
quality_counts = df['quality'].value_counts()

plt.figure(figsize=(8, 6))
plt.bar(quality_counts.index, quality_counts, color='deeppink')
plt.title('CountPlot of Quality')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

Measuring skewness of the data reveals the symmetry of the distribution.  For a value near 0, the data set is symmetric about its mean while for values 1 or larger, the data is positive (right) skewed,  meaning the data has a long tail to the right with very large values of the independent variable.

```python
# kernel density plot shows variance in data set
sns.set_style('darkgrid') # dark background with white grid lines, easier to read

numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns # identifies all columns with numerical data type, returns column names

plt.figure(figsize=(14, len(numerical_columns) * 3))
for idx, feature in enumerate(numerical_columns, 1): # make sure to start at 1 and not 0, since matplotlib indexes starting at 1
    plt.subplot(len(numerical_columns), 2, idx) # plot has (num_rows, num_cols, plot_index)
    sns.histplot(df[feature], kde=True) # kde = Kernel density estimate (smooth curve) added over bars of histogram
    plt.title(f'{feature} | Skewness: {round(df[feature].skew(), 2)}') # skewness is a measure of the asymmetry of the distribution

plt.tight_layout()
plt.show()
```
