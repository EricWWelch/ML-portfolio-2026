Basic linear regression model using randomly generated data, following https://www.geeksforgeeks.org/machine-learning/ml-linear-regression/

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(42)

X = np.random.rand(50, 1) * 100
Y = 3.5 * X + np.random.randn(50, 1) * 20
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
Y_pred = model.predict(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X, Y, color='b', label='Data Points')
plt.plot(X, Y_pred, color='r', linewidth=2, label='Regression Line')
plt.title('Linear Regression on Random Dataset')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f'The slope is m = {model.coef_[0][0]:.3f}, and the intercept is b = {model.intercept_[0]:.3f}')
print(f'The form of the line is y = {model.coef_[0][0]:.3f}x + {model.intercept_[0]:.3f}')
```

This shows the functional relationship between the independent variable
